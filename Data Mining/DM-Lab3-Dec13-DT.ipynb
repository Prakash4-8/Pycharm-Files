```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
import pandas as pd
data = pd.read_csv("C:/Users/bablu/Desktop/Data Set/golf-dataset.csv")
data.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Outlook'] = le.fit_transform(data['Outlook'])
data['Temp'] = le.fit_transform(data['Temp'])
data['Humidity'] = le.fit_transform(data['Humidity'])
data['Windy'] = le.fit_transform(data['Windy'])
data['Decision'] = le.fit_transform(data['Decision'])
```

```python
#split dataset in features and target variable
feature_cols = ['Outlook', 'Temp', 'Humidity', 'Windy']
X = data[feature_cols] # Features
y = data['Decision'] # Target variable
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=20) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#!pip install matplotlib
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
